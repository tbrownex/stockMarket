```python
import pandas as pd
from datetime import datetime
```

```python
loc = '~/data/stockMarket/'

yahoo = ['spx', 'DAX', 'russel_2000', 'Nikkei225']
MSCI = ['MSCI World', 'MSCI World - US', 'EAFE']
```

```python
def cleanYahoo(idx):
    df = pd.read_csv(loc + idx +'.csv', thousands=',')
    df=df[['Date', 'Close']]
    df=df[:-1]
    df.columns=['date', 'value']
    df['date'] = df['date'].apply(lambda x: pd.to_datetime(x))
    df.to_csv(loc + '/clean/' + idx +'.csv', index=False)
```

```python
def cleanMSCI(idx):
    df = pd.read_csv(loc + idx +'.csv', thousands=',')
    df.columns=['date', 'value']
    df=df[:613]
    df['date'] = df['date'].apply(lambda x: pd.to_datetime(x))
    df.to_csv(loc + '/clean/' + idx +'.csv', index=False)
```

```python
for idx in yahoo:
    print(idx)
    cleanYahoo(idx)
```

```python
for idx in MSCI:
    print(idx)
    cleanMSCI(idx)
```
