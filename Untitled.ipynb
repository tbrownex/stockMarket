```python
import pandas as pd
import matplotlib.pyplot as plt
import glob
```

##### Identify indexes that are anti-correlated to SPX

```python
def getFilenames():
    fileNames = []
    for file in glob.glob("/home/tbrownex/data/stockMarket/clean/*.csv"):
        fileNames.append(file)
    return fileNames
```

```python
def formatName(fullPath):
        csv = fullPath.split("/")[-1]
        return csv.split(".")[0]
```

```python
def printFilenames(fileNames):
    for f in fileNames:
        idx = formatName(f)
        print(idx)
```

```python
loc = '~/data/stockMarket/clean/'

spx = pd.read_csv(loc + 'spx.csv', parse_dates=['date'])
```

```python
fileNames = getFilenames()
# printFilenames(fileNames)
```

```python
print("{:35}{}".format('Index', 'Correlation'))
for idx in fileNames:
    idxData = pd.read_csv(idx, parse_dates=['date'])
    startDt = idxData['date'].min()
    tmp = spx[spx['date'] >= startDt]
    tmp = tmp.reset_index()
    assert len(tmp) == len(idxData), "series length mismatch"
    corr = idxData['value'].corr(tmp.value)
    idx = formatName(idx)
    print("{:<30}{:.2f}".format(idx, corr))
```

```python
fig = plt.figure(figsize=(17,4))
plt.plot(tmp.value, c='r')
plt.plot(idxData.value, c='g')
plt.show()
plt.close()
```
